## Plot relaxation over value function

```python
from src.analyses.convergence_analysis import HausdorffAnalyzer
from src.utility.plot import plot_converge_order
from NSPLIB.src.instances.tanksize.tanksize import const_model
from src.models.cz_model import CaoZavalaModel, CaoZavalaAlgo
from src.models.lagrangean_model import LagrangeanModel, LagrangeanAlgo
import numpy as np
import matplotlib.pyplot as plt
from pyomo.opt import SolverFactory
import copy
sto_m = const_model()
m = CaoZavalaModel.from_sto_m(sto_m)
m.build()
y_val_fix={'productTankSize[1]': 676.3117608642578,
  'productTankSize[2]': 541.855562210083,
  'productTankSize[3]': 257.58401122093204}
eps=1
steps=10
idx="productTankSize[1]"
idx_interval=np.linspace(y_val_fix[idx]-eps, y_val_fix[idx]+eps, steps)
val_func_obj=[]
aux_model=copy.deepcopy(m.aux_models["lbd"][0])
for idx_val in idx_interval:
    y_val_fix[idx]=idx_val
    for y_idx in y_val_fix:
        aux_model.y[y_idx].fix(y_val_fix[y_idx])
    result=SolverFactory("baron").solve(aux_model)
    val_func_obj.append(result['Problem'][0]['Lower bound'])
```

```python
val_func_obj
```

```python
interval={i:(y_val_fix[i],y_val_fix[i]) for i in y_val_fix}
interval[idx]=(y_val_fix[idx]-eps, y_val_fix[idx]+eps)
sto_m = const_model()
m = LagrangeanModel.from_sto_m(sto_m)
m.build()
alg = LagrangeanAlgo(m,lag_iter=50, solver='baron')
from src.models.bb_node import BranchBoundNode
from src.analyses.convergence_analysis import HausdorffAnalyzer 
node=BranchBoundNode(interval)
lbd,mu_set, obj_set=alg.calc_lbd(node,tol=1e-2)
```

```python
relaxed_func=[]
size1_interval=np.linspace(interval[idx][0], interval[idx][1], steps)
for k in size1_interval:
    interval["productTankSize[1]"]=(k,k)
    relaxed_func.append(obj_set[-1][0] + sum([interval[idx][0] * mu_set[-1][0][idx] for idx in interval]))
```

```python
plt.plot(size1_interval,val_func_obj)
plt.plot(size1_interval,relaxed_func,color='red')
```
